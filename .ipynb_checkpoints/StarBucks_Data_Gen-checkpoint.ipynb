```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
raw_data = pd.read_csv('static/assets/raw_starbucks.csv')
raw_data
```

```python
sizes = {
    'Short' : 1,
    'Tall' : 2,
    'Grande' : 3,
    'Venti' : 4,
}

def find_size(prep):
    # print(f'prep: {prep}')
    for size in sizes:
        # print(size, size in prep)
        if size in prep:
            return size
    return None

def remove_size(prep):
    re = ''
    s = find_size(prep)
    if s:
        re = prep.replace(s, '').strip()
    # print(s, " - ", re, " - ", prep)
    if re != '':
        return re
    return prep

size_data = []
cur_size = None
for prep in raw_data['Beverage_prep']:
    s = find_size(prep)
    if s:
        cur_size = s
    size_data.append(cur_size)

def remove_perc(s):
    return float(s.replace('%', '').strip())

clean_data = raw_data.copy()
clean_data = clean_data.rename(columns={col: col.strip() for col in clean_data.columns})
clean_data.insert(3, 'Size', size_data)
clean_data = clean_data.assign(Beverage_prep = clean_data['Beverage_prep'].apply(remove_size))
for col in ['Vitamin A (% DV)', 'Vitamin C (% DV)', 'Calcium (% DV)', 'Iron (% DV)']:
    clean_data = clean_data.assign(**{col: clean_data[col].apply(remove_perc)})
clean_data
```

```python
def filter_prep(s):
    if s in ['Nonfat Milk', '2% Milk', 'Soymilk', 'Solo', 'Doppio']:
        return False
    else:
        return True
export_df = clean_data[clean_data['Beverage_prep'].apply(filter_prep)]
export_df = export_df[['Beverage_category', 'Beverage', 'Size', 'Calories','Total Fat (g)', 'Protein (g)', 'Caffeine (mg)']]
# export_df = clean_data
export_df
```

```python
export_df.to_csv('static/assets/starbucks.csv', index=False)
```
